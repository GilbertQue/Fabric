```python
# Parameters
# NOTE: Make sure to attach the notebook to one of the new Schema Enabled Lakehouses
source_workspace_name = "Fabric - FourMoo"
source_lakehouse_name = "LH_NewSchema"
source_schema_name = "dbo"
source_table_name = "source_salesoutput"

destination_workspace_name = "FourMoo-Fabric-Demo"
destination_lakehouse_name = "Dest_LH_NewSchema"
destination_schema_name = "staging"
destination_table_name = "tb salesOutput"
```

```python
# Create the Four Part naming for the Source and Destination Tables
source_full_table_name = f"`{source_workspace_name}`.`{source_lakehouse_name}`.`{source_schema_name}`.`{source_table_name}`"
destination_full_table_name = f"`{destination_workspace_name}`.`{destination_lakehouse_name}`.`{destination_schema_name}`.`{destination_table_name}`"

# Check the outputs are correct
print(source_full_table_name)
print(destination_full_table_name)
```

```python
# Define the insert query
# NOTE: The destination Table has already been created.
insert_query = f"""
    insert into {destination_full_table_name}
    Select *    
    from {source_full_table_name}

"""

#Write Data
spark.sql(insert_query)
```
